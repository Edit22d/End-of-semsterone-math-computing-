## iii) A HISTOGRAM 

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from numpy import random
poisson_data = random.poisson(lam = 5,size=1000)
sns.histplot(poisson_data,label='poisson',kde = 'False',color = 'purple')
plt.title('A HITOGRAM SHOWING  THE SAMPLES OF POISSON DISTRIBUTION',color ='green')
plt.xlabel('Density',color= 'green')
plt.ylabel('Numbers', color='green')
plt.show()
```
